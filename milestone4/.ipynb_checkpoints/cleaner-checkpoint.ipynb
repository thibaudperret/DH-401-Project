```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('../../survey of music taste（回复） (1).xlsx', )
```

```python
data = data.drop(data.columns[[0, -1]], axis=1).replace(np.nan, "")
```

```python
clean_data = pd.DataFrame()
```

```python
for i, col in enumerate(data.columns):
    print(i, '->', col)
```

```python
clean_data['Age'] = data['Age']
clean_data['Gender'] = data['Gender']
clean_data['Classical'] = data['The average time  [Classical]']
clean_data['Rock'] = data['The average time  [Rock]']
clean_data['Pop'] = data['The average time  [Pop]']
clean_data['R&B/Soul'] = data['The average time  [R&B and Soul]']
clean_data['Electronic'] = data['The average time  [Electronic]']
clean_data['Folk'] = data['The average time  [Folk]']
clean_data['Jazz'] = data['The average time  [Jazz]']
clean_data['Blues'] = data['The average time  [Blues]']
clean_data['Hiphop'] = data['The average time  [Hip hop]']
clean_data['Total'] = data['How many hours per week do you spend listening to the musical genre you prefer (the case(s) with the highest scores in the previous question)']
clean_data['School'] = data['which university do you belong to?']
```

```python
section = [3, 19]
degree = [4, 7, 9, 11, 13, 15, 17, 20, 22, 24, 26, 28, 30, 32]
major = [5, 6, 8, 10, 12, 14, 16, 18, 21, 23, 25, 27, 29, 31, 33]
```

```python
clean_data['Section'] = data[data.columns[section[0]]]
clean_data['Degree'] = data[data.columns[degree[0]]]
clean_data['Major'] = data[data.columns[major[0]]]

for table, name in [(section, 'Section'), (degree, 'Degree'), (major, 'Major')]:
    for col in table[1:]:
        clean_data[name] += data[data.columns[col]]
```

```python
clean_data.to_csv('clean.csv')
```
